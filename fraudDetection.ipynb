```python
#train
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
import os

print("1. Loading datasets...\n")
train_df = pd.read_csv('phishing_train.csv')
test_df = pd.read_csv('phishing_test.csv')

print(f'train samples: {train_df.shape[0]} samples')
print(f'test samples: {test_df.shape[0]} samples')

print('training distribution:')
print(train_df['Label'].value_counts())

#test distribution
print(test_df['Label'].value_counts())

X_test_raw = train_df['URL']
X_test_raw = test_df['URL']
y_train = train_df['Label']
y_test = test_df['Label']

# Convert labels to binary (0 for good, 1 for bad)
y_train_binary = (y_train == 'bad').astype(int)
y_test_binary = (y_test == 'bad').astype(int)

print("\n2. Training TF-IDF + Logistic Regression Model...")
tfidf_lr_pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(
        analyzer = 'char',
        ngram_range = (2,4),
        max_features = 10000,
        lowercase = True,
        sublinear_tf = True
        
    )),
    ('clf', LogisticRegression(
        C=1.0,
        max_iter = 1000,
        random_state  = 42,
        class_weight = 'balanced'
    ))
])

print("   Training...")

tfidf_lr_pipeline.fit(X_train, y_train_binary)

# Evaluate on test set
y_pred_lr = tfidf_lr_pipeline.predict(X_test_raw)#Makes a final decision for each URL: Is it malicious or not?

y_pred_proba_lr = tfidf_lr_pipeline.predict_proba(X_test_raw)[:, 1]#Gives a confidence score for each URL: How sure is the model?

print("\n3. Training TF-IDF + Random Forest Model...")
tfidf_rf_pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(
        analyzer='char',
        ngram_range=(2, 4),
        max_features=5000,  # Smaller feature set for RF
        lowercase=True,
        sublinear_tf=True
    )),
    ('clf', RandomForestClassifier(
        n_estimators=100,
        max_depth=20,
        min_samples_split=5,
        random_state=42,
        class_weight='balanced',
        n_jobs=-1
    ))
])

```
